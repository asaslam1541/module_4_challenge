```python
# Convert the 'date' column to a DateTimeIndex
whale_returns_df['Date'] = pd.to_datetime(whale_returns_df['Date'])

# Set the 'date' column as the DataFrame's index
whale_returns_df.set_index('Date', inplace=True)
```

```python
# Convert to DateTimeIndex
sp_tsx_history_df["Date"] = pd.to_datetime(sp_tsx_history_df["Date"])

# Set index to 'Date'
sp_tsx_history_df.set_index("Date", inplace=True)

# Print
sp_tsx_history_df
```

```python
# To identify null values
whale_returns_df.isnull()

# Remove rows wit null values
whale_returns_df.dropna(inplace=True)

# Remove columns with null values (in-place)
whale_returns_df.dropna(axis=1, inplace=True)

# Print
whale_returns_df
```

```python
# To identify null values
algo_returns_df.isnull()

# Remove rows wit null values
algo_returns_df.dropna(inplace=True)

# Remove columns with null values (in-place)
algo_returns_df.dropna(axis=1, inplace=True)

# Print
algo_returns_df
```

```python
# To identify null values
sp_tsx_history_df.isnull()

# Remove rows wit null values
sp_tsx_history_df.dropna(inplace=True)

# Remove columns with null values (in-place)
sp_tsx_history_df.dropna(axis=1, inplace=True)

# Print
sp_tsx_history_df
```

```python
sp_tsx_history_df['Close'] = sp_tsx_history_df['Close'].str.replace('$', '')
```

```python
sp_tsx_history_df['Close'] = sp_tsx_history_df['Close'].str.replace(',', '')
```

```python
sp_tsx_history_df
```

```python
# Convert 'Close' to numeric
sp_tsx_history_df['Close'] = pd.to_numeric(sp_tsx_history_df['Close'], errors='coerce')
```

```python
sp_tsx_history_df
```

```python
tsx_daily_returns = sp_tsx_history_df.pct_change()
```

```python
tsx_daily_returns.dropna()
```

```python
combined_df = pd.concat([whale_returns_df,algo_returns_df, tsx_daily_returns], axis='columns', join='inner')
```

```python
combined_df
```

```python
# Rename the column from 'Close' to 'S&P TSX'
combined_df.rename(columns={'Close': 'S&P TSX'}, inplace=True)

# Print
combined_df
```

```python
# Calculate cumulative returns
cumulative_returns = (1 + combined_df).cumprod()
# Plot cumulative returns
cumulative_returns.plot(figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.title('Cumulative Returns of Portfolios and S&P TSX 60')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create box plots for returns
combined_df.boxplot(figsize=(18, 12))
plt.title('Box Plots of Portfolio Returns')
plt.ylabel('Returns')
plt.grid(True)
plt.show()
```

```python
# Calculate standard deviation
portfolio_std = combined_df.std()
tsx_std = combined_df['S&P TSX'].std()

# print results
print (portfolio_std)
print(tsx_std)
```

```python
# Determine portfolios riskier than S&P TSX 60
riskier_portfolios = portfolio_std[portfolio_std > tsx_std].index.tolist()

# Print result
print(riskier_portfolios)
```

```python
# Calculate annualized standard deviation
annualized_std = portfolio_std * (252 ** 0.5)
```

```python
# Print results
print('Standard Deviation:')
print(portfolio_std)
print('\nPortfolios Riskier than S&P TSX 60:')
print(riskier_portfolios)
print('\nAnnualized Standard Deviation:')
print(annualized_std)
```

```python
# Calculate rolling standard deviation with a 21-day window
rolling_std = combined_df.rolling(window=21).std()
```

```python
# Calculate correlation between each portfolio and S&P TSX 60
correlation_matrix = combined_df.corr()
```

```python
# Choose one portfolio
chosen_portfolio = 'BERKSHIRE HATHAWAY INC'
```

```python
# Calculate rolling beta between the chosen portfolio and S&P TSX 60 with a 60-day window
rolling_covariance = combined_df[chosen_portfolio].rolling(window=60).cov(combined_df['S&P TSX'])
rolling_variance = combined_df['S&P TSX'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
```

```python
# Plot rolling standard deviation
rolling_std.plot(figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Rolling Standard Deviation')
plt.title('Rolling Standard Deviation of Portfolios and S&P TSX 60')
plt.legend()
plt.grid(True)
plt.show()

# Plot correlation matrix
import seaborn as sns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Plot rolling beta
rolling_beta.plot(figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Rolling Beta')
plt.title('Rolling Beta between ' + chosen_portfolio + ' and S&P TSX 60')
plt.grid(True)
plt.show()
```

```python
# Calculate daily returns for risk-free rate (e.g., a 10-year government bond yield)
risk_free_rate = 0.01 / 252  # Replace with the appropriate risk-free rate

# Calculate the excess returns of portfolios over the risk-free rate
excess_returns = combined_df - risk_free_rate

# Calculate the annualized Sharpe ratio for each portfolio
sharpe_ratios = (excess_returns.mean() * 252) / (excess_returns.std() * (252 ** 0.5))
```

```python
# Plot Sharpe ratios in a bar plot
plt.figure(figsize=(12, 6))
sharpe_ratios.plot(kind='bar', color='blue', alpha=0.7)
plt.xlabel('Portfolio')
plt.ylabel('Sharpe Ratio')
plt.title('Sharpe Ratios of Portfolios and S&P TSX 60')
plt.grid(axis='y')
plt.show()
```
